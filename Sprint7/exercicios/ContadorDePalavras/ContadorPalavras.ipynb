```python
from pyspark import SparkContext

if 'sc' in locals():
    sc.stop()

sc = SparkContext("local", "WordCount")

arquivo = "README.md" 

text_file = sc.textFile(arquivo)

ContarPalavras = text_file.flatMap(lambda linha: linha.split(" ")) \
                           .map(lambda palavra: (palavra, 1)) \
                           .reduceByKey(lambda a, b: a + b)

for palavra, count in ContarPalavras.collect():
    print(f"{palavra}: {count}")
```
